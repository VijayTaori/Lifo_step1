```python
df_eur_buy['Amount_open_Eur'] = df_eur_buy['Amount']/df_eur_buy['Low_x']
df_eur_buy['Open Rate_Eur'] = df_eur_buy['Open Rate']/df_eur_buy['Low_x']
df_eur_buy['Close Rate_Eur'] = df_eur_buy['Close Rate']/df_eur_buy['High_y']
df_eur_buy['Spread_Eur'] = df_eur_buy['Spread']/df_eur_buy['Low_y']
df_eur_buy['Fee_Eur'] = df_eur_buy['Rollover Fees and Dividends']/df_eur_buy['Low_y']
df_eur_buy['diff_rates'] = df_eur_buy['Close Rate_Eur'] - df_eur_buy['Open Rate_Eur']
df_eur_buy['profit_1_eur'] = df_eur_buy['diff_rates'] * df_eur_buy['Units']
df_eur_buy['real_profit_eur'] = df_eur_buy['profit_1_eur'] + df_eur_buy['Fee_Eur'] - df_eur_buy['Spread_Eur']
df_eur_buy['Close_amount_eur'] = df_eur_buy['real_profit_eur'] + df_eur_buy['Amount_open_Eur']
df_eur_buy['Profit_Diff'] = df_eur_buy['real_profit_eur'] - df_eur_buy['Profit']
```

```python
df_eur_buy_profit_sort = df_eur_buy.sort_values(by='Profit_Diff')
```
